## Removing Outliers

Now we will remove some outliers to increase the performance of our model

```python
# Outlier Removal
def remove_outliers(feature, fname, data):
  # Identify 25th and 75th quartile
  q25, q75 = np.percentile(feature, 25), np.percentile(feature, 75)
  # Finding inter quartile range
  iqr = q75 - q25
  # Defining upper and lower bounds for outliers
  upper = q75 + 1.5*iqr
  lower = q25 - 1.5*iqr

  # Finding outliers
  outliers = [x for x in feature if x < lower or x > upper]
  print(fname + f' : {len(outliers)}')

  # Dropping outliers
  df = data.drop(data[(data[fname] > upper) | (data[fname] < lower)].index)

  return df

for col in data:
  new_data = remove_outliers(data[col], str(col), data)
```

## Dividing data

```python
# Dividing data into features and labesl

X = new_data.drop(['Bankrupt?'], axis=1) # Features
y = new_data['Bankrupt?'] # Labels
```

```python
# Normalising the data using logarithmic transformation

def log_trans(data):
  for col in data:
    skew = data[col].skew()
    if skew > 0.5 or skew < -0.5:
      data[col] = np.log1p(data[col])
    else:
      continue

  return data

norm_X = log_trans(X)
```

```python
# Stratified Cross Validation Splitting
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)

for train_index, test_index in skf.split(X_train, y_train):
  X_train_skf, X_val_skf = X_train.iloc[train_index], X_train.iloc[test_index]
  y_train_skf, y_val_skf = y_train.iloc[train_index], y_train.iloc[test_index]


X_train_skf = X_train_skf.values
X_val_skf = X_val_skf.values
y_train_skf = y_train_skf.values
y_val_skf = y_val_skf.values

```

### Logistic Regression

```python


acc_reg = []
precision_reg = []
recall_reg =[]
f1_reg = []
auc_reg = []

log_reg = LogisticRegression(random_state=42,
                             max_iter=1000)

log_reg_params = {"penalty": ['l2'],
                  'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
                  'class_weight': ['balanced',None],
                  'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}

rs_log_reg = RandomizedSearchCV(log_reg,
                                param_distributions=log_reg_params,
                                n_iter=4)

for train, val in skf.split(X_train_skf, y_train_skf):
  pipeline_reg = imb_make_pipeline(SMOTE(sampling_strategy='minority'),
                                   rs_log_reg)
  reg_model = pipeline_reg.fit(X_train_skf[train], y_train_skf[train])
  best_est_reg = rs_log_reg.best_estimator_
  preds_reg = best_est_reg.predict(X_train_skf[val])

  acc_reg.append(reg_model.score(X_train_skf[val], y_train_skf[val]))
  precision_reg.append(precision_score(y_train_skf[val], preds_reg))
  recall_reg.append(recall_score(y_train_skf[val], preds_reg))
  f1_reg.append(f1_score(y_train_skf[val], preds_reg))
  auc_reg.append(roc_auc_score(y_train_skf[val], preds_reg))



```

```python
print('Logistic Regression (SMOTE) results:\n')
print(f'accuracy : {np.mean(acc_reg)}')
print(f'precision : {np.mean(precision_reg)}')
print(f'recall : {np.mean(recall_reg)}')
print(f'f1_score : {np.mean(f1_reg)}')

# y_preds = org_log_reg.predict(X_test)
# print(classification_report(y_test, y_preds))
```

```python
# Printing the classification report

label = ['Fin.Stable', 'Fin.Unstable']
pred_reg_skf = best_est_reg.predict(X_val_skf)
print(classification_report(y_val_skf, pred_reg_skf, target_names=label))
```

```python
# Plotting Precision-Recall curve

y_score_reg = best_est_reg.predict(X_val_skf)

average_precision = average_precision_score(y_val_skf, y_score_reg)

fig = plt.figure(figsize=(12,6))

precision, recall, _ = precision_recall_curve(y_val_skf, y_score_reg)

plt.step(recall, precision, color='r', alpha=0.2,
         where='post')
plt.fill_between(recall, precision, step='post', alpha=0.2,
                 color='#F59B00')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('OverSampling Precision-Recall curve: \n Average Precision-Recall Score ={0:0.2f}'.format(
          average_precision), fontsize=15)

plt.show()
```

```python
from sklearn.metrics import roc_curve

log_reg_fpr, log_reg_tpr, log_reg_threshold = roc_curve(y_val_skf, pred_reg_skf)
rfc_fpr, rfc_tpr, rfc_threshold = roc_curve(y_val_skf, pred_rfc_skf)
xgb_fpr, xgb_tpr, xgb_threshold = roc_curve(y_val_skf, pred_xgb_skf)
cat_fpr, cat_tpr, cat_threshold = roc_curve(y_val_skf, pred_cat_skf)

def graph_roc_curve_multiple(log_reg_fpr, log_reg_tpr, rfc_fpr, rfc_tpr,xgb_fpr, xgb_tpr,cat_fpr, cat_tpr):
    plt.figure(figsize=(20,8))
    plt.title('ROC Curve', fontsize=14)
    plt.plot(log_reg_fpr, log_reg_tpr, label='Logistic Regression Classifier Score: {:.4f}'.format(roc_auc_score(y_val_skf, pred_reg_skf)))
    plt.plot(xgb_fpr, xgb_tpr, label='XGBoost Classifier Score: {:.4f}'.format(roc_auc_score(y_val_skf, pred_xgb_skf)))
    plt.plot(cat_fpr, cat_tpr, label='CatBoost Classifier Score: {:.4f}'.format(roc_auc_score(y_val_skf, pred_cat_skf)))
    plt.plot(rfc_fpr, rfc_tpr, label='Random Forest Classifier Score: {:.4f}'.format(roc_auc_score(y_val_skf, pred_rfc_skf)))
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([-0.01, 1, 0, 1])
    plt.xlabel('False Positive Rate', fontsize=13)
    plt.ylabel('True Positive Rate', fontsize=13)
    plt.annotate('Minimum ROC Score of 50% \n (This is the minimum score to get)', xy=(0.5, 0.5), xytext=(0.6, 0.3),
                arrowprops=dict(facecolor='#6E726D', shrink=0.05),
                )
    plt.legend()
    
graph_roc_curve_multiple(log_reg_fpr, log_reg_tpr, rfc_fpr, rfc_tpr,xgb_fpr, xgb_tpr,cat_fpr, cat_tpr)
plt.show()
```
